```python
def get_chemical_values_over_time(state_name, county_names=None):
    df = chemicals_df[chemicals_df['state'] == state_name]
    all_counties = df['county'].unique()
    if county_names:
        random_counties = np.array(county_names)
    else:
        random_counties = np.random.choice(all_counties, 5)
    counties = { name: None for name in random_counties }
    chemicals = list(df['chemical_species'].unique())
    for county in counties:
        county_chemicals = df[df['county'] == county]
#         values = [[] for _ in range(len(chemicals))]
        chems_final = defaultdict(lambda: {})
        for i,chemical in enumerate(chemicals):
            curr_chems = county_chemicals[county_chemicals['chemical_species'] == chemical]
            years = curr_chems['year'].unique()
            values = []
            for year in years:
                vals_in_year = curr_chems[curr_chems['year'] == year]['value'].values
                values.append(np.mean(vals_in_year))
            chems_final[chemical] = (years, values)  
        counties[county] = dict(chems_final)
    return counties
```

```python
output = get_chemical_values_over_time('CA')
```

```python
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
import copy
```

```python
with_nitrates = copy.deepcopy(output)
for county in output:
    del output[county]['Nitrates']
```

```python
def output_to_plots(output):
    counties = {}
    for county in output:
        data = []
        chemicals = output[county]
        for chemical in chemicals:
            years,values = chemicals[chemical]
            trace = go.Scatter(
                x = years,
                y = values,
                name = chemical
            )
            data.append(trace)
        counties[county] = data
    return counties
```

```python
county_data = output_to_plots(output)
county_data_with_nitrates = output_to_plots(with_nitrates)
```

```python
def plot_county(county):
    layout = dict(title = 'Chemical Contaminate Levels in ' + county + ' by Year',
              xaxis = dict(title = 'micrograms/L'),
              yaxis = dict(title = 'Year'),
              )
    fig = dict(data=county_data[county], layout=layout)
    py.iplot(fig, filename=county)
```
